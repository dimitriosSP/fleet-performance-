```python
import pandas as pd
import glob
from pandas import read_excel
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
import plotly.graph_objects as go
import seaborn as sns
```

```python
# File paths for each dataset

path_sfoc = r'C:\Users\dps\Documents\FLEET PERFORMANCE\2.Sfoc_calibration/Phoebe2.xlsx'
path_power = r'C:\Users\dps\Documents\FLEET PERFORMANCE\2.Sfoc_calibration\Phoebe1.xlsx'

# Load the datasets

df_sfoc = pd.read_excel(path_sfoc, decimal=',', parse_dates = ['Date'] )
df_power = pd.read_excel(path_power, decimal=',', parse_dates = ['Date'] )

```

```python
# Test that datasets are loaded properly (to do so we call the first 2 lines of the dataset)

df_sfoc.head(2)
```

```python
# Merge the dataset

merged = pd.merge(df_power, df_sfoc, on="Date")
```

```python
# Keep only 'True' values

merged_True = merged[(merged['Visible_power'] == 'True') & (merged['Visible_sfoc'] == 'True') ]
```

```python
# Correlation for all variables of the dataset

merged_True.corr()
```

```python
print(merged_True.Date.min())
print(merged_True.Date.max())
```

```python
# POWER - SFOC plot

df = merged_True

fig = go.Figure(go.Scatter(
    x = df['Corrected Power'],
    y = df['SFOC (%)'],
    mode="markers"
))

fig.update_layout(title="POWER - SFOC       Period: 6/2021- 6/2022")

fig.show()
```

```python
# Comments on the above chart

# we assume negative correlation when SFOC values are high.
# For example, when SFOC is around 104 - 106, POWER is around 75 - 100

# For some other areas we observe zero correlation
```

# Split Dataset to intervals

```python
# groupby your key and freq

g = merged_True.groupby(pd.Grouper(key='Date', freq='M'))
```

```python
# groups to a list of dataframes with list comprehension

dfs = [group for _,group in g]
```

```python
# Split the Dataset to each month

for i in range (0, 12):
    print('Month:', i, dfs[i].corr())
```

```python
print('first day:', dfs[8].Date.min())
print('last day:', dfs[8].Date.max())
```

```python
import plotly.express as px
import pandas as pd

df = merged_True

fig = px.scatter(df, x='Date', y=['SFOC (%)', 'Corrected Power'])

fig.update_traces(
    name='Corrected Power',
    mode='markers+lines',
    error_y=dict(color='black', width=5, thickness=1),
    marker=dict(color='red', size=10),
    line=dict(color='blue', width=3),
    showlegend=True
)

fig.update_layout(template='simple_white',
    yaxis_title='Corrected Power & SFOC',
    title='Continuous, variable value error bars',
    hovermode="x"
)
 
fig.show()
```
